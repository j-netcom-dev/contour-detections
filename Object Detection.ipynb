```python
from IPython import display
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from matplotlib.image import imread
import warnings, numpy as np, cv2, math
warnings.filterwarnings("ignore")
```

```python
def segment_image(mat: np.ndarray):
    shape =mat.shape
    color_channels =shape[-1]
    image =mat.astype('float32') /255.
    X_input =image.reshape(-1, color_channels)
    kmeans =KMeans(n_clusters =color_channels).fit(X_input)
    segmented_image =kmeans.cluster_centers_[kmeans.labels_]
    return segmented_image.reshape(shape)
```

```python
def plot_images(image1:np.ndarray, image2:np.ndarray, image3:np.ndarray):
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 3, 1) 
    plt.imshow(image1)
    plt.title('Original')
    plt.subplot(1, 3, 2) 
    plt.imshow(image2)
    plt.title('Segment')
    plt.subplot(1, 3, 3) 
    plt.imshow(image3)
    plt.title('Contours')
    plt.tight_layout()
    plt.show()
```

```python
def canny_edge_detection(im_mat:np.ndarray):
    gray = cv2.cvtColor((im_mat *255).astype('uint8'), cv2.COLOR_RGB2GRAY)
    blur =cv2.GaussianBlur(gray, (5, 5), 1)
    canny =cv2.Canny(blur, threshold1 =50, threshold2 =255)
    return canny
```

```python
cone1 =imread('cone1.jpg')
cone1_segmented =segment_image(cone1)
cone1_edges =canny_edge_detection(cone1_segmented)
plot_images(cone1, cone1_segmented, cone1_edges)
```
